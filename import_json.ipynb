```python
import pymysql, os, json

 
arquivo_ura = os.path.abspath('../../..') + "/UraTelefonica.json"
json_data_ura=open(arquivo_ura).read()
json_obj_ura = json.loads(json_data_ura)

arquivo_chat = os.path.abspath('../../..') + "/ChatOnline.json"
json_data_chat=open(arquivo_chat).read()
json_obj_chat = json.loads(json_data_chat)

def validate_string(val):

   if val != None:
        if type(val) is int:
            #for x in val:
            #   print(x)
            return str(val).encode('utf-8')
        else:
            return val

con = pymysql.connect(host = 'localhost',user = 'root',passwd = '123456',db = 'busclick')
cursor = con.cursor()

for i, item in enumerate(json_obj_ura):

    Data = validate_string(item.get("data", None))
    Time = validate_string(item.get("Hor√°rio", None))
    fila_descricao = validate_string(item.get("fila_descricao", None))
    Agente = validate_string(item.get("Agente", None))
    Status = validate_string(item.get("Status", None))
    Telefone = validate_string(item.get("telefone", None))
    Estado = validate_string(item.get("Estado", None))
    email = validate_string(item.get("email", None))
    fila_de_atendimento = validate_string(item.get("fila_de_atendimento", None))
    duracao_espera_fila = validate_string(item.get("duracao_espera_fila", None))
    duracao_atendimento = validate_string(item.get("duracao_atendimento", None))
    duracao_total_da_chamada = validate_string(item.get("duracao_total_da_chamada", None))
    Classificao = validate_string(item.get("classificao", None))
    Sub_classificao = validate_string(item.get("sub_classificao", None))

    cursor.execute("INSERT INTO tb_UraTelefonica (Data, Time, fila_descricao, Agente, Status, Telefone, Estado, email, fila_de_atendimento, duracao_espera_fila, duracao_atendimento, duracao_total_da_chamada, Classificao, Sub_classificao ) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)", (Data,        Time, fila_descricao, Agente, Status, Telefone, Estado, email, fila_de_atendimento, duracao_espera_fila, duracao_atendimento, duracao_total_da_chamada, Classificao, Sub_classifica))

               

for i, item in enumerate(json_obj_chat):

    dt_data_inicio = validate_string(item.get("Data da conversa (Inicio)", None))
    dt_data_fim = validate_string(item.get("Data da conversa (Fim)", None))
    Semana = validate_string(item.get("Semana", None))
    Agente = validate_string(item.get("Agente", None))
    email = validate_string(item.get("Visitor_Email", None))
    feedback_response_additional = validate_string(item.get("feedback_response_additional", None))
    feedback_average = validate_string(item.get("feedback_average", None))
    Atendido = validate_string(item.get("Atendido", None))

    cursor.execute("INSERT INTO tb_chatonline (dt_data_inicio, dt_data_fim, Semana, Agente, email, feedback_response_additional, feedback_average, Atendido ) VALUES (%s, %s, %s, %s, %s, %s, %s, %s)", (dt_data_inicio, dt_data_fim, Semana, Agente, email, feedback_response_additional, feedback_average, Atendido ))   

con.commit()
con.close()
```
